```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch


fig, ax = plt.subplots(1, 1, figsize=(10, 10))

data2010 = gpd.read_file('AGEB_urb_2010_5.shp')

data2010["ENT"] = data2010['CVEGEO'].str.slice(0,2)
data2010["MUN"] = data2010["CVEGEO"].str.slice(2,5)

datayucatan2010 = data2010[data2010['ENT'] == '31']
merida = datayucatan2010[datayucatan2010["MUN"]=="050"]

datayucatan2020 = gpd.read_file('31a.shp')

merida2020 = datayucatan2020[datayucatan2020["CVE_MUN"] == "050"]


merida_to_set = set(merida['CVEGEO'])
merida_2_toset = set(merida2020['CVEGEO'])
remained = merida_to_set.intersection(merida_2_toset)


new = merida_2_toset - merida_to_set
eliminated = merida_to_set - merida_2_toset

remained = merida[merida['CVEGEO'].isin(remained)]
new = merida2020[merida2020['CVEGEO'].isin(new)]
eliminated = merida[merida['CVEGEO'].isin(eliminated)]

plt.grid()

remained.plot(ax=ax, color='blue', edgecolor='black')
new.plot(ax=ax, color='red', edgecolor='black')
eliminated.plot(ax=ax, color='yellow', edgecolor='black')


plt.show()
```
